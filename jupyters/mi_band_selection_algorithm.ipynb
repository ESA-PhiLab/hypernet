<font size="5">Mutual Information-based band selection algorithm</font>
=
Mutual information, (MI) measures the statistical dependence between two random variables, (here, spectral bands and ground truth map can be treated as discrete random variables), thus, MI can evaluate the relative utility of each band to classification.

In order to run MI-based band selection algoritm, number of arguments need to be specified:
* `data_path` - Path to the dataset, which can be either a *matlab* or *numpy* file.
* `ref_map_path` - Path to the reference map, i.e. the ground truth map, which can be either a *matlab* or *numpy* file.
* `dest_path` - Destination path of the directory designed for storing artifacts.
* `bands_num` - Number of bands to select.
* `eta` - Parameter referred in the paper as *complementary threshold*. This argument is dataset dependent.
* `bandwidth` - Parameter referred in the paper as *rejection bandwidth*. This argument is dataset dependent.

[Link to paper](https://ieeexplore.ieee.org/document/1715309)

```python
import os

DATA_DIR = os.getenv("HYPERNET_DATA_DIR", os.path.join("..", "..", "hypernet-data"))
RESULTS_DIR = os.path.join(
    os.getenv("HYPERNET_RESULTS_DIR", os.path.join("..", "..", "hypernet-data", "results")),
    "mi_band_selection"
)
```

```python
from python_research.experiments.band_selection_algorithms.mi.mi_band_selection import Arguments, main

arguments = Arguments(
    data_path=os.path.join(DATA_DIR, "PaviaU_corrected.npy"),
    ref_map_path=os.path.join(DATA_DIR, "PaviaU_gt.npy"),
    dest_path=RESULTS_DIR,
    bands_num=11,
    eta=0.01,
    bandwidth=3
)

main(args=arguments)
```
